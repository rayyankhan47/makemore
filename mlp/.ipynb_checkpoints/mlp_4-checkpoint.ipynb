# This notebook continues from `mlp_3.ipynb`, and we will now start performing certain experiments in scaling our NN in order to stop underfitting. The two experiments are to:
1. Increase hidden layer size
2. Increase embedding dimension (increase m in R^m)

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = open("names.txt", "r").read().splitlines()
words[:10]
```

```python
# build vocab of chars and mappings to/from ints in alphabetical order
chars = sorted(list(set("".join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi["."] = 0
itos = {i:s for s,i in stoi.items()}
itos
```

```python
# build the dataset

def build_dataset(words):
    block_size = 3
    X, Y = [], []
    for w in words:
        context = [0] * block_size
        for ch in w + ".":
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]

    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27,2), generator=g)
W1 = torch.randn((6,100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
sum(p.nelement() for p in parameters)
```

```python
for p in parameters:
    p.requires_grad = True
```

```python
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre
```

```python
#lri = []
lossi = []
stepi = []

for i in range(40000):

    # minibatch
    ix = torch.randint(0, Xtr.shape[0], (32,))

    # forward pass
    emb = C[Xtr[ix]]
    h = torch.tanh(emb.view(-1,6) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])

    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    # update
    lr = 0.05 # changed from 0.1 to 0.05 near when I felt we were approaching a bottleneck
    for p in parameters:
        p.data += -lr * p.grad

    # stats 
    #lri.append(lre[i])
    stepi.append(i)
    lossi.append(loss.item())

print(loss.item())
```

```python
plt.plot(stepi, lossi)
plt.show()
```

```python
emb = C[Xdev]
h = torch.tanh(emb.view(-1,6) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits, Ydev)
print(loss.item())
```

## Visualization of embeddings in R^2
Let's visualize what embedding locations look like in R^2 before we increase m.

## Changing some hyperparameters
Let's change the embedding dimension and the number of neurons in the first layer.

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27,10), generator=g)
W1 = torch.randn((30,200), generator=g)
b1 = torch.randn(200, generator=g)
W2 = torch.randn((200,27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
sum(p.nelement() for p in parameters) # significantly higher now w/ increased factors
```

```python
for p in parameters:
    p.requires_grad = True
```

```python
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre
```

```python
# lri = []
lossi = []
stepi = []
```

```python
for i in range(50000):

    # minibatch
    ix = torch.randint(0, Xtr.shape[0], (32,))

    # forward pass
    emb = C[Xtr[ix]]
    h = torch.tanh(emb.view(-1,30) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])

    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    # update
    lr = 0.1
    for p in parameters:
        p.data += -lr * p.grad

    # track stats
    # lri.append(lre[i])
    stepi.append(i)
    lossi.append(loss.log10().item()) # .log10() to squash things for better vis

print(loss.item())
```
