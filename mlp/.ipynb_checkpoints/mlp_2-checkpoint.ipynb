# This notebook does what was done in `mlp_1.ipynb`, but with all words from the dataset this time.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = open("names.txt", "r").read().splitlines()
words[:10]
```

```python
# build vocab of chars and mappings to/from ints in alphabetical order
chars = sorted(list(set("".join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi["."] = 0
itos = {i:s for s,i in stoi.items()}
itos
```

```python
# build the dataset

block_size = 3 # context length
X, Y = [], []
for w in words: # all words

    # print("=======================")
    # print(w)
    context = [0] * block_size # padding
    for ch in w + ".":
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        # print("".join(itos[i] for i in context), "--->", itos[ix])
        context = context[1:] + [ix] # crop and append

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
X.shape, Y.shape # dataset
```

```python
g = torch.Generator().manual_seed(2147483647) # this is for reproducibility
C = torch.randn((27,2), generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100,27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
sum(p.nelement() for p in parameters) # number of total parameters
```

```python
for p in parameters:
    p.requires_grad = True
```

## Minibatching
By having added minibatching, doing 10, even 100 iterations went from taking
3-4 seconds to being instantaneous. We are maximizing efficiency on parallelism. 

Also, by having added minibatching, the quality of our gradient has gotten a bit worse, so we aren't in the perfect gradient direction. However, it's better to have a good enough gradient and take many steps than to have a perfect gradient and take fewer steps. That's why minibatching works quite well in practice.

```python
"""
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre

lri = []
lossi = []

for i in range(1000):
    
    # minibatch
    ix = torch.randint(0, X.shape[0], (32,))
    # 32 just happens to be a good common batch size.
    # Also, X.shape[0] is the number of examples overall in words.
    
    # forward pass
    emb = C[X[ix]] # (32, 3, 2)
    h = torch.tanh(emb.view(-1,6) @ W1 + b1) # (32, 100)
    logits = h @ W2 + b2 # (32,27)
    loss = F.cross_entropy(logits, Y[ix])

    # print(loss.item())
    
    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    # update
    lr = lrs[i]
    for p in parameters:
        p.data += -lr * p.grad

    # track stas for learning rate
    lri.append(lr)
    lossi.append(loss.item())

print(loss.item())
"""
```

## Learning Rate
We have no idea as to whether -0.1 is too large or too small of a learning rate in our gradient descent. One way to do figure something out here is to vary the learning rate's absolute value from 0.001 to 1 and to plot that against its respective obtained losses, and to find a nice learning rate that decreases loss the best. So we replace `for _ in range(100)` with `for i in range(100)` and use that index to go over the varying learning rates from 0.001 to 1.

```python
plt.plot(lri, lossi)
plt.show()
```

## A good learning rate is...
0.1, as it turns out. That's actually precisely what we had when we started...well anyway, now we can be sure that 0.1 is actually a good decision.

Below, let's rewrite everything, but with the correct learning rate.

```python
for i in range(10000):
    
    # minibatch
    ix = torch.randint(0, X.shape[0], (32,))
    # 32 just happens to be a good common batch size.
    # Also, X.shape[0] is the number of examples overall in words.
    
    # forward pass
    emb = C[X[ix]] # (32, 3, 2)
    h = torch.tanh(emb.view(-1,6) @ W1 + b1) # (32, 100)
    logits = h @ W2 + b2 # (32,27)
    loss = F.cross_entropy(logits, Y[ix])

    # print(loss.item())
    
    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    # update
    lr = 0.1 if loss.item() > 2.36 else 0.01 # the lr drops in a magnitude once we start to plateau in loss. I chose 2.36 by observing.
    for p in parameters:
        p.data += -lr * p.grad

print(loss.item())
```

## Splitting the dataset
Right now, even though our loss has decreased quite a bit, comparably to what we had with the bigram model, the model is still not good, because if we were to sample the model's results, we would get results essentially identical to the dataset, which is not what we're looking for.
Typically what we do is we split the dataset into training, validation, and test, so that we can avoid underfitting and overfitting.
